```python
d={(i,j,k):((i*j*k)/2>i+j+k) for i in range(1,7) for j in range(1,7) for k in range(1,7)}

dinv = defaultdict(list)

for i,j in d.items():
  dinv[j].append(i)

for key,val in dinv.items():
  print(str(key)+":"+str(val))
  X=len(val)/216
  print(str(key)+":"+str(X))
```
