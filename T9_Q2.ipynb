```python
# prompt: store a list of x,y variables without using loop

import numpy as np
from sklearn.metrics import mean_squared_error

x = np.array([1, 0.8, 2, 2.5, 3, 4, 4.2, 6, 6.3, 7, 8, 8.2, 9])
y = np.array([2, 3, 2.5, 1, 2.3, 2.8, 1.5, 2.6, 3.5, 4, 3.5, 5, 4.5])


```

```python
y1 = []
y2 = []

for i in range(len(x)):
  if x[i] <= 5:
    y2.append(y[i])
  else:
    y1.append(y[i])

print("y1:", y1)
print("y2:", y2)

y1_mean=np.mean(y1)*np.ones(len(y1))
y2_mean=np.mean(y2)*np.ones(len(y2))

#print("y1_mean:", y1_mean)
#print("y2_mean:", y2_mean)

y1_mse=mean_squared_error(y1, y1_mean)
y2_mse=mean_squared_error(y2, y2_mean)

print("y1_mse:", y1_mse)
print("y2_mse:", y2_mse)
```

```python
#overall MSE at depth 1

overall_mse=(len(y1)*y1_mse+len(y2)*y2_mse)/len(y)
print("overall_mse at depth 1:", overall_mse)

#finding MSE at root

y_mean=np.mean(y)*np.ones(len(y))
root_mse=mean_squared_error(y, y_mean)
print("root_mse:", root_mse)
```
