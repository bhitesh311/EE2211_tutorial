```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import inv
```

```python
df = pd.read_csv("government-expenditure-on-education.csv")
expenditureList = df ['recurrent_expenditure_total'].tolist()
yearList = df ['year'].tolist()
```

```python
#The given system is over-determined as there are more equations than unknown
m_list = [[1]*len(yearList), yearList] #adding bias term to the matrix for obtaining linear regresion equation
X = np.array(m_list).T
y = np.array(expenditureList)
w = inv(X.T @ X) @ X.T @ y
print(w)
y_line = X.dot(w)
plt.plot(yearList, expenditureList, 'o', label = 'Expenditure over the years')
plt.plot(yearList, y_line,label='linear regression')
plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Expenditure')
plt.title('Education Expenditure')
plt.legend()
plt.show()
```

```python
#making prediction for the year 2021
y_predict = np.array([1, 2021]).dot(w)
print(y_predict)
```
