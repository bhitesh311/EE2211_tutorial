```python
import numpy as np
from numpy.linalg import inv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.array([[1,-10],[1,-8],[1,-3],[1,-1],[1,2],[1,8]]) #accomodating for the bias
y=np.array([[5],[5],[4],[3],[2],[2]])
P=np.array([[1,-10,100,-1000],[1,-8,64,-512],[1,-3,9,-27],[1,-1,1,-1],[1,2,4,8],[1,8,64,512]])

#polynomial regression
w_poly=inv(P.T@P)@P.T@y
print(w_poly)

#linear regression
w_lin=inv(x.T@x)@x.T@y
print(w_lin)

#curve equations
y_linear=x@w_lin
y_poly=P@w_poly

#sketching the result
plt.scatter(x[:,1],y,label=' training samples')
plt.plot(x[:,1],y_linear,color='orange',label='linear regression')
plt.plot(x[:,1],y_poly,color='grey',label='Polynomial regression')

# Highlight X and Y axes separately
plt.axhline(y=0, color='black', linewidth=1.5)  # X-axis (horizontal line)
plt.axvline(x=0, color='black', linewidth=1.5)  # Y-axis (vertical line)

plt.legend()
plt.show()
```

Q 2b) prediction for x=9

```python
x_test=np.array([[1,9]])
P_new = np.array([[1,9,81,729]])

y_linear_predict=x_test@w_lin
y_poly_predict=P_new@w_poly

print(y_poly_predict)
print(y_linear_predict)
```
