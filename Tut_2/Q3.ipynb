```python
import pandas as pd 
from sklearn.datasets import load_iris
from pandas.plotting import scatter_matrix
```

```python
iris_dataset=load_iris()
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split( iris_dataset['data'],iris_dataset['target'], random_state=0)
#random state ensures that everytime you run the code you get the same output.
#X_train --> training set features
#X_test --> testing set features
#y_train --> training set labels
#y_test --> testing set labels
```

```python

iris_dataframe  = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
#iris_dataset.feature_names column names for the dataset. Ex: sepal length, sepal width etc.
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20})
#c=y_train --> colors the points based on their class labels
# bins : Defines the number of equal-width bins in the range
```
